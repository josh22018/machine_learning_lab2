```python
from sklearn.model_selection import train_test_split
purchase_data['Category'] = np.where(purchase_data['Payment (Rs)'] > 200, 'RICH', 'POOR')

# Feature Engineering: Extract relevant features
# For example, you might use columns like 'Product_1', 'Product_2', 'Product_3', etc.

# Select features and target variable
features = purchase_data[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']]  # Adjust based on your features
target = purchase_data['Category']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Choose a classifier (Random Forest as an example)
classifier = RandomForestClassifier()

# Train the model
classifier.fit(X_train, y_train)

# Make predictions
y_pred = classifier.predict(X_test)

# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```
